```python
df_news = pd.read_csv('./data/news.csv', encoding='cp949')
# df_news.head()
df_news=df_news['news']
# print(df_news)
text=' '.join(df_news)
print(text)
# print(df_news[df_news.category==7][news])
# text = ' '.join(df_news[df_news.category == 6].news.values)
# print(text)
```

```python
import re

cleaned_text = re.sub('[^\w\s]','',text)
cleaned_text = re.sub('\n',' ',cleaned_text)
```

```python
from konlpy.tag import Okt
tagger = Okt()

noun = tagger.nouns(cleaned_text)
noun2more = [ele for ele in noun if len(ele) > 1]
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec([noun2more],   # 리스트 형태의 데이터
                  sg=1,         # 0: CBOW, 1: Skip-gram
                  vector_size=100,     # 벡터 크기
                  window=3,     # 고려할 앞뒤 폭(앞뒤 3단어)
                  min_count=3)  # 사용할 단어의 최소 빈도(3회 이하 단어 무시)
```

```python
sim = model.wv.most_similar('대회')
sim
```

```python
sim = model.wv.similarity('대회', '올림픽')
print(sim)
```

```python
sim = model.wv.most_similar('아이폰')
print(sim)
```

```python
from gensim.models import FastText
f_model = FastText([noun2more], sg=1, vector_size=100, window=3, min_count=3)
```

```python
f_model.wv.most_similar('대회')
```

```python
sim = f_model.wv.similarity('대회', '올림픽')
print(sim)
```

```python
f_model.wv.most_similar('아이폰')
```
